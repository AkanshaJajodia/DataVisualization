```python
print('importing libraries')
import pandas as pd
import requests
from bs4 import BeautifulSoup 
print('libraries imported ')
```

```python
start_url = 'https://en.wikipedia.org/wiki/Tesla,_Inc.'

#downloading HTML from URL
download_html = requests.get(start_url)

# parse HTML using soup object

soup = BeautifulSoup(download_html.text)
#print (soup.prettify())
# save a local copy
with open('download_html','w') as file:
    file.write(soup.prettify())
```

```python
# scrapping table 
f_table = soup.select('table.wikitable')
print(f_table)
```

```python
f_tabletbdy = soup.select('table.wikitable tbody')[0]  # 0 to select the first element of list
print(f_tabletbdy)
```

```python
# extracting table headings
table_head = soup.select('table.wikitable tbody tr th')
print(table_head)

```

```python
table_columns = []

for element in table_head:
    col_label = element.get_text(separator =' ', strip = True)
    table_columns.append(col_label)
    print("Col is : ", col_label)
print(table_columns)
```

```python
# replacing spaces with _

table_columns = []

for element in table_head:
    col_label = element.get_text(separator =' ', strip = True)
    col_label = col_label.replace(' ','_')
    table_columns.append(col_label)
    print("Col is : ", col_label)
print(table_columns)
```

```python
# using regular exp to get rid of [b]
import re
regex = re.compile('_\[\w\]') # looks for _ followd by []
table_columns = []

for element in table_head:
    col_label = element.get_text(separator =' ', strip = True)
    col_label = col_label.replace(' ','_')
    col_label= regex.sub('',col_label)
    table_columns.append(col_label)
    
print(table_columns)
```

```python
# extracting rows
table_rows = soup.select('table.wikitable tbody tr' )
#print(table_rows)
table_data =[]
for index, element in enumerate (table_rows):
    if index >0:
        row_list =[]
        values = element.select('td')
        for value in values:
            row_list.append(value.text)
        table_data.append(row_list)  
print(table_data)
```

```python
# getting rid of \n
table_data =[]
for index, element in enumerate (table_rows):
    if index >0:
        row_list =[]
        values = element.select('td')
        for value in values:
            row_list.append(value.text.strip())
        table_data.append(row_list)  
print(table_data)

```

```python
# creating pandas df
df = pd.DataFrame(table_data
                 )
df
```
